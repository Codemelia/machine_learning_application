```python
from nltk import word_tokenize
import nltk

text = 'he likes cats and dogs, and teaching machines to learn'
```

# **Stemming**

```python
from nltk.stem import SnowballStemmer
```

```python
stem = SnowballStemmer(language='english')
print([stem.stem(token) for token in word_tokenize(text)])
```

# **Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
```

```python
#nltk.download('stopwords')
nltk.download('wordnet')
```

```python
lm = WordNetLemmatizer()
print([lm.lemmatize(token, 'n') for token in word_tokenize(text)])
```

```python
lemmatization_result = []
for token in word_tokenize(text):
    lemmatization_result.append(lm.lemmatize(token, 'v'))
print(lemmatization_result)
```
