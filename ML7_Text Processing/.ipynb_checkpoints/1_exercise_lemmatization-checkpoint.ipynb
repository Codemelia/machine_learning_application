## Mini Exercise 1: Lemmatization

1.	We have the following list of short text documents:

    docs = [
        "The cat sat on the mat",
        "The dog sat on the log",
        "Cats and dogs are friends",
        "The mat was sat on by the cat"
    ]

2.	There is no need to remove stop words or punctuation. However, lemmatization, which reduces words to their root form, is required. 

3.	Manually perform Bag-of-Words (BOW) on the documents and display the feature vector for each document. 

4.	Write code to perform Bag-of-Words (BOW) on the documents and display the feature vector for each document. 

5.	Verify that the results from Steps 3 and 4 are the same.

```python
# Imports

import string
import pandas as pd
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Init nltk resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Define docs
docs = [
    "The cat sat on the mat",
    "The dog sat on the log",
    "Cats and dogs are friends",
    "The mat was sat on by the cat"
]
```

```python
# Lemmatization function
def lemmatize(docs):
    lemmatized = []

    # Get lemmatizer object from NLTK
    lemma = WordNetLemmatizer()

    # Iterate docs
    for doc in docs:
        # Convert all words to lowercase
        words = doc.lower().split()

        # Lemmatize to root form
        words = [ lemma.lemmatize(word, pos = 'v') for word in words ]

        # Join each word to form document
        lemmatized.append(' '.join(words))

    return lemmatized
```

```python
# Format vectors and indexes
def pretty_print(feat_vecs, feat_names):
    df = pd.DataFrame(
        data = feat_vecs,
        index = [ 'doc' + str(i + 1) for i in range(len(docs)) ],
        columns = feat_names
    )

    print(df)
```

```python
# BOW function
def BOW(docs):
    bow = CountVectorizer()

    # Fit > learn vocab from docs
    # Transform > Convert docs to vectors
    # toarray() > transform results to dense matrix form
    feature_vectors = bow.fit_transform(docs).toarray()

    # Return both feature vectors and feature names
    return feature_vectors, bow.get_feature_names_out()
```

```python
print(f"Before lemmatization:\n{docs}")

# Lemmatize docs
lemmatized = lemmatize(docs)
print(f"After lemmatization:\n{lemmatized}")
```

```python
# Perform BOW
feat_vecs, feat_names = BOW(lemmatized)
print("After BOW:\n")
pretty_print(feat_vecs, feat_names)
```
