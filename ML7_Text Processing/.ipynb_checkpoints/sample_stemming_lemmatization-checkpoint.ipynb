```python
import nltk
from nltk import word_tokenize

text = 'he likes cats and dogs, and teaching machines to learn'
```

# **Stemming**

```python
from nltk.stem import SnowballStemmer
nltk.download('punkt_tab')
```

```python
stem = SnowballStemmer(language='english')
print([stem.stem(token) for token in word_tokenize(text)])
```

# **Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
```

```python
#nltk.download('stopwords')
nltk.download('wordnet')
```

```python
lm = WordNetLemmatizer()
print([lm.lemmatize(token, 'n') for token in word_tokenize(text)])
```

```python
lemmatization_result = []
for token in word_tokenize(text):
    lemmatization_result.append(lm.lemmatize(token, 'v'))
print(lemmatization_result)
```

# **Stop Words**

```python
from nltk.corpus import stopwords
import string
```

```python
stops = stopwords.words('english')
print('no. of stop words: ', len(stops))
```

```python
print(stops)
```

```python
# With removing punctuation
punc = str.maketrans('', '', string.punctuation)
text_no_punc = text.translate(punc)

terms = [token for token in lemmatization_result if token not in stops and token in text_no_punc]

# Without removing punctuation
# terms = [token for token in lemmatization_result if token not in stops]

print(terms)
```

```python
# With removing punctuation
terms = [token for token in word_tokenize(text_no_punc) if token in stops]

# Without removing punctuation
# terms = [token for token in lemmatization_result if token in stops]
print(terms)
```
