## Part 1 — Series: Labelled 1D Arrays

A **Series** is a 1D labelled array — think of it as a single column in a spreadsheet. Each element has an **index label** (default: 0, 1, 2… or custom names). Values are **mutable**, but the size is **immutable** after creation.

| Accessor | Behaviour |
|---|---|
| `.iloc[n]` | Integer position (0-based) |
| `.loc['label']` | Label name |
| `iloc[a:b]` slicing | **Exclusive** — returns a, a+1, … b-1 |
| `loc['a':'b']` slicing | **Inclusive** — returns a … b |

````python
import pandas as pd

# Default integer index
s1 = pd.Series([21, 32, 45])
print(f"Default index:\n{s1}\n")

# Custom named index
s2 = pd.Series([21, 32, 45],
               index=['mary', 'john', 'james'])
print(f"Named index:\n{s2}\n")

# Access by position (iloc) vs label (loc)
print(f"iloc[1] = {s2.iloc[1]}")
print(f"loc['john'] = {s2.loc['john']}")

# CRITICAL: loc slicing is INCLUSIVE!
print(f"\niloc[0:2]:\n{s2.iloc[0:2]}")  # mary, john
print(f"\nloc['mary':'john']:\n{s2.loc['mary':'john']}")  # mary, john (inclusive!)

# Update value
s2['john'] = 38
print(f"\nAfter update:\n{s2}")
````

## Part 2 — DataFrame: The Core Data Structure for ML

A **DataFrame** is a 2D labelled table — rows × columns. Each column is a Series sharing the same index. DataFrames are the primary structure for ML: features are columns, observations are rows.

````python
import pandas as pd

# Method 1: From dictionary (most common!)
df = pd.DataFrame({
    'name': ['Mary', 'John', 'James'],
    'age': [21, 32, 45],
    'height': [1.7, 1.8, 1.6],
    'weight': [55, 65, 50]
})
print(f"From dict:\n{df}\n")

# Method 2: From list of lists + column names
df2 = pd.DataFrame(
    [[21, 1.7, 55], [32, 1.8, 65], [45, 1.6, 50]],
    columns=['age', 'height', 'weight'],
    index=['mary', 'john', 'james']
)
print(f"From lists + custom index:\n{df2}\n")

# Key attributes
print(f"Shape: {df2.shape}")
print(f"Columns: {list(df2.columns)}")
print(f"Index: {list(df2.index)}")
print(f"Dtypes:\n{df2.dtypes}")
````

## Part 3 — Selecting, Filtering & Indexing Data

Pandas offers multiple ways to select data. The two primary accessors are:
- **`iloc`** — integer **i**ndex position
- **`loc`** — **l**abel-based access

Boolean filtering lets you extract rows that satisfy a condition — the foundation of data querying.

````python
import pandas as pd

df = pd.DataFrame({
    'age': [21, 32, 45],
    'height': [1.7, 1.8, 1.6],
    'weight': [55, 65, 50]
}, index=['mary', 'john', 'james'])
print(f"DataFrame:\n{df}\n")

# Column selection
print(f"df['age'] (Series):\n{df['age']}\n")
print(f"df[['age','weight']] (DataFrame):")
print(f"{df[['age','weight']]}\n")

# Row by position (iloc)
print(f"iloc[1] (Series):\n{df.iloc[1]}\n")
print(f"iloc[[1]] (DataFrame):\n{df.iloc[[1]]}\n")

# Row by label (loc)
print(f"loc['john','age'] = {df.loc['john','age']}")

# Slicing rows + cols
print(f"\niloc[1:, 1:]:\n{df.iloc[1:, 1:]}")
````

### Boolean Filtering

Filter rows by applying conditions — similar to SQL `WHERE`.  
**Important**: when combining conditions, always use parentheses and `&` / `|` (not `and` / `or`).

````python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice','Bob','Charlie','Diana','Eve'],
    'age': [25, 30, 35, 28, 42],
    'dept': ['Eng','Sales','Eng','HR','Sales'],
    'salary': [70000, 55000, 80000, 48000, 62000]
})
print(f"All data:\n{df}\n")

# Simple filter
print(f"Age > 30:\n{df[df['age'] > 30]}\n")

# Combined (MUST use parentheses!)
mask = (df['age'] > 25) & (df['dept'] == 'Eng')
print(f"Age>25 AND Eng:\n{df[mask]}\n")

# isin — match against a list
print(f"Dept in Sales/HR:")
print(df[df['dept'].isin(['Sales', 'HR'])])
````

## Part 4 — Data Wrangling: Modifying, Cleaning & Transforming

Real-world data is always messy. Data wrangling is the process of cleaning, reshaping, and preparing data for analysis or modelling. This typically takes **70–80% of a data scientist's time**.

### Adding & Modifying Columns

````python
import pandas as pd

df = pd.DataFrame({
    'name': ['Mary', 'John', 'James'],
    'height': [1.7, 1.8, 1.6],
    'weight': [55, 65, 50]
})
print(f"Original:\n{df}\n")

# Add computed column (feature engineering!)
df['bmi'] = df['weight'] / df['height']**2
print(f"With BMI:\n{df}\n")

# Categorise with apply + lambda
df['status'] = df['bmi'].apply(
    lambda x: 'Normal' if x < 25 else 'Overweight'
)
print(f"With status:\n{df}\n")

# Rename columns
df = df.rename(columns={'name': 'patient'})

# Drop column
df = df.drop(columns=['status'])
print(f"After rename + drop:\n{df}")
````

### Handling Missing Data

Missing values (`NaN`) are common in real datasets. Always **detect** first, then decide: drop or impute.

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'age': [25, np.nan, 35, 28],
    'salary': [50000, 60000, np.nan, 48000]
})
print(f"With missing values:\n{df}\n")

# Detect missing values
print(f"NaN counts:\n{df.isnull().sum()}\n")

# Option 1: Drop rows with any NaN
print(f"After dropna():\n{df.dropna()}\n")

# Option 2: Fill with column mean (impute)
df_filled = df.copy()
df_filled['age'] = df_filled['age'].fillna(
    df_filled['age'].mean()
)
df_filled['salary'] = df_filled['salary'].fillna(
    df_filled['salary'].median()
)
print(f"After imputation:\n{df_filled}")
````

### Sorting & Ranking

````python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice','Bob','Charlie','Diana','Eve'],
    'age': [25, 30, 35, 28, 42],
    'salary': [70000, 55000, 80000, 48000, 62000]
})

# Sort by salary (descending)
print(f"By salary (desc):")
print(df.sort_values('salary', ascending=False))

# Top 3 earners
print(f"\nTop 3 earners:")
print(df.nlargest(3, 'salary'))

# Sort by multiple columns
print(f"\nBy age then salary:")
print(df.sort_values(['age', 'salary']))
````

## Part 5 — GroupBy, Aggregation & Merging

GroupBy implements the **split-apply-combine** pattern: split data into groups, apply an aggregation function, and combine results. Merging joins two DataFrames like SQL JOIN.

### GroupBy — Split-Apply-Combine

````python
import pandas as pd

df = pd.DataFrame({
    'dept': ['Eng','Sales','Eng','HR','Sales','HR'],
    'name': ['Alice','Bob','Charlie','Diana','Eve','Frank'],
    'salary': [70000, 55000, 80000, 48000, 62000, 52000]
})
print(f"Data:\n{df}\n")

# Average salary per department
print(f"Mean salary by dept:")
print(df.groupby('dept')['salary'].mean())

# Multiple aggregations
print(f"\nMultiple stats:")
print(df.groupby('dept')['salary'].agg(
    ['mean', 'min', 'max', 'count']
))

# Group size
print(f"\nEmployees per dept:")
print(df.groupby('dept').size())
````

### Merging & Concatenating

Pandas `merge` is equivalent to SQL JOIN. `concat` stacks DataFrames vertically or horizontally.

````python
import pandas as pd

# Two related tables
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'dept_id': [10, 20, 10, 30]
})

departments = pd.DataFrame({
    'dept_id': [10, 20, 40],
    'dept_name': ['Engineering', 'Sales', 'Marketing']
})

# Inner join (default) — only matching rows
inner = pd.merge(employees, departments, on='dept_id')
print(f"Inner join:\n{inner}\n")

# Left join — keep all employees
left = pd.merge(employees, departments,
                on='dept_id', how='left')
print(f"Left join:\n{left}\n")

# Concat — stack vertically
batch1 = pd.DataFrame({'x': [1,2], 'y': [3,4]})
batch2 = pd.DataFrame({'x': [5,6], 'y': [7,8]})
combined = pd.concat([batch1, batch2], ignore_index=True)
print(f"Concat:\n{combined}")
````

### Apply — Custom Transformations

`apply` lets you run any function over rows or columns. `map` replaces values using a dictionary. These are the tools for custom feature engineering.

````python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'score': [85, 62, 91],
    'grade_letter': ['A', 'B', 'A']
})
print(f"Original:\n{df}\n")

# Apply lambda to transform a column
df['pass_fail'] = df['score'].apply(
    lambda x: 'Pass' if x >= 70 else 'Fail'
)

# Map categorical values to numbers
df['grade_num'] = df['grade_letter'].map(
    {'A': 4, 'B': 3, 'C': 2}
)

# Apply to entire row (axis=1)
df['summary'] = df.apply(
    lambda row: f"{row['name']}: {row['score']}",
    axis=1
)

print(f"Transformed:\n{df}")
````
