## Workshop 1 - Text Summarisation

Objective: To practise using TFIDF as tool for text featurisation

Exercise

You are given a text file called “space_invaders.txt”. Your task is to produce a 10-line
summary of the text.
Here is an approach you can consider:

• First, compute the TFIDF value of each word in the text file by treating each
sentence as a document and the entire text file as the corpus.

• Then, sum the TFIDF values of each sentence.

• Finally, selects sentences that have the top 10 TFIDF sums and print them in the
order that they appear in the text file.
For pre-processing, remember to remove stop-words from your corpus and stem each
word in the text.

```python
# Imports
import string
import pandas as pd
import numpy as np
import nltk

from nltk import word_tokenize
from nltk import sent_tokenize
from nltk import pos_tag

from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Pretty print function
def pretty_print(feat_vecs, feat_names):
    df = pd.DataFrame(data=feat_vecs,
            index=['doc' + str(idx) for idx in range(len(feat_vecs))],
            columns=feat_names)

    print(df)
```

```python
# STEP 0: Read text file as corpus
# Read text file
def read_text(txt):

    # Read file as text
    with open(txt, 'r', encoding = 'utf-8') as f:
        text = f.read()

    # Tokenise text into sentences and return as corpus
    return sent_tokenize(text)

docs = read_text('space_invaders.txt')
print(docs[:5]) # Check split
```

```python
# STEP 1: Compute TFIDF value of each word in each document

# Docs cleaning function
def preprocess(docs):
    clean_docs = []

    # Get stemmer from NLTK
    stem = SnowballStemmer(language = 'english')

    # Get list of stopwords
    stop_words = stopwords.words('english')

    # Mapper to replace punctuations
    punc = str.maketrans('', '', string.punctuation)

    # Implement for each document
    for doc in docs:
        # A. Remove punctuation and conv to lowercase
        doc_no_punc = doc.translate(punc)
        words = doc_no_punc.lower().split()

        # B. Stem and remove stop words
        words = [
            stem.stem(token)
                for word in words if word not in stop_words
                for token in word_tokenize(word)
        ]

        # C. Join each word in list to form document
        clean_docs.append(' '.join(words))

    return clean_docs

# Compute tfidf value of each word function
def compute_tfidf(docs):
    tfidf = TfidfVectorizer()
    feat_vecs = tfidf.fit_transform(docs).toarray()
    return feat_vecs, tfidf.get_feature_names_out()

# Preprocess data
clean_docs = preprocess(docs)
print(clean_docs[:5])

# Compute TFIDF values
feat_vecs, feat_names = compute_tfidf(clean_docs)
pretty_print(feat_vecs, feat_names)
```

```python
# STEP 2. Sum TFIDF values of each sentence
# Summing function
def sum_tfidf(feat_vecs, feat_names):
    df = pd.DataFrame(
        feat_vecs,
        index = [ (idx + 1) for idx in range(len(feat_vecs)) ],
        columns = feat_names
    )

    tfidf_sums = df.sum(axis = 1).to_frame(name = 'TFIDF_Sum')
    tfidf_sums.index.name = 'Document'

    return tfidf_sums

tfidf_sums = sum_tfidf(feat_vecs, feat_names)
print(tfidf_sums)
```

```python
# STEP 3. Select sentences that have top 10 tfidf sums, 
# print in order that they appear in text file

print(tfidf_sums.nlargest(10, columns = 'TFIDF_Sum').sort_index())
```
