```python
# Imports
import nltk
import numpy as np
import string
import pandas as pd
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

1. First, treat each quote as a separate document and compute TFIDF values for each word in the 51 documents (remember to treat each line of quote as a separate document; your corpus is the 51 documents).

```python
# Read text file
def read_text(path):
    file = open(path, encoding = 'utf-8')
    text = file.read()
    docs = text.split('\n')
    return docs

docs = read_text('quotes.txt')
print(len(docs)) # Check split
```

```python
# Pre processor
def preprocess(docs):
    # Use english stop words from NLTK
    stop_words = stopwords.words('english')

    # Use stemmer to convert words to root forms
    stemmer = nltk.stem.PorterStemmer()

    clean_docs = []

    # Remove punctuation
    punc = str.maketrans('', '', string.punctuation)

    # Implement stemming and conv each word to lowercase
    for doc in docs:
        doc_no_punc = doc.translate(punc)
        words_stemmed = [ stemmer.stem(w)
            for w in doc_no_punc.lower().split()
            if w not in stop_words ]
        clean_docs.append(' '.join(words_stemmed))

    return clean_docs
```

```python
# Pretty print
def pretty_print(feat_vecs, feat_names):
    df = pd.DataFrame(
        data = feat_vecs,
        index = [ idx for idx in range(len(feat_vecs)) ],
        columns = feat_names
    )
    print(df)

# Compute tfidf value of each word
def compute_tfidf_docs(docs):
    feat_vecs = tfidf.fit_transform(docs).toarray()
    return feat_vecs, tfidf.get_feature_names_out()

# Get tfidf obj
tfidf = TfidfVectorizer()

# Clean and fit + transform docs
clean_docs = preprocess(docs)
docs_vecs, docs_names = compute_tfidf_docs(clean_docs)
pretty_print(docs_vecs, docs_names)
```
