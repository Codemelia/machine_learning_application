```python
# write code to standardize the given dataset and show that, the standardized 
# dataset has 0 mean and a standard deviation of 1.

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# reads datasets

df = pd.read_csv("mini_data.csv")
print(f"df={df}")
```

```python
# perform standardization
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df)

# after scaling
print(f"\nAfter standardization: {x_scaled}")
```

```python

# show that mean = 0
mean = np.mean(x_scaled, axis=0).round()
print(f"\nMean = {mean}")

# show that std dev = 1
std = np.std(x_scaled, axis=0)
print(f"Std Dev = {std}")

```
