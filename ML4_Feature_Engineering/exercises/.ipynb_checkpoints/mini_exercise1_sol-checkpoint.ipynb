```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# =========================================
# (1) Load CSV file
# =========================================

df = pd.read_csv("bike_types_times.csv")

# Show first few rows
print(df.head())
```

```python

# =========================================
# (2) One-hot encode feature "Road_Bike_Type"
# =========================================

# Create encoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform categorical column
encoded = encoder.fit_transform(df[["Road_Bike_Type"]])

# Convert encoded result into DataFrame
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(["Road_Bike_Type"])
)

# Combine with original numerical columns
df_ohot = pd.concat(
    [df[["Price", "Time_100m_s"]], encoded_df],
    axis=1
)

# Show result
print("\nOne-hot encoded dataframe:")
print(df_ohot.head())
```
