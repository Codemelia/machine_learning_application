```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('iris-data-clean.csv')
# exclude frist row, exclude first and last column
# first row is header, first column is number series, last column is category
x = data.iloc[1:, 1:-1]

print(x.head())
```

```python
# use StandardScaler to zero-mean
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# apply PCA, number of components is 2
pca = PCA(n_components=2)
new_features = pca.fit_transform(x_scaled)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```
