1. The file mini_data.csv contains 3 features (X, Y and Z) and the table below shows our entire
dataset.

```python
# Load csv file
mini_data = pd.read_csv('mini_data.csv')
print(mini_data)
```

2. The dataset has not been standardized.
3. Write Python code to perform standardization on the data, and show that your standardized data (i.e. each of your non-categorical features) has zero-mean and unit-variance. Feel free to use any NumPy functions, such as numpy.mean() and numpy.std(), in your solution.

```python
# 1. Use StandardScaler to zero-mean
scaler = StandardScaler()
mini_data_scaled = scaler.fit_transform(mini_data)
print(f"After standardisation:\n{mini_data_scaled}")

# 2. Show that mean = zero (zero-mean)
# Round to show zero-values (if mean is close to zero)
mean = np.mean(mini_data_scaled, axis = 0).round()
print(f"\nMean: {mean}")

# 3. Show that SD (variance**2) = 1, meaning variance = 1 (unit-variance)
sd = np.std(mini_data_scaled, axis = 0)
print(f"\nStandard deviation: {sd}")
```
