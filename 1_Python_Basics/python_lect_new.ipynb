## Collection Data Types


### Lists — Mutable Ordered Collections


```python
# Creating a list
students = ["Alice", "Bob", "Charlie"]
print("Original:", students)

# Append and insert
students.append("Diana")
students.insert(1, "Eve")
print("After add:", students)

# Access by index (0-based!)
print(f"First: {students[0]}")
print(f"Last:  {students[-1]}")

# Modify in place (mutable!)
students[0] = "Alicia"
print("Modified:", students)

# List of mixed types
record = ["John", 36, 1.75, True]
print(f"\nMixed list: {record}")
print(f"Length: {len(record)}")
```

### Tuples — Immutable Ordered Collections


```python
# Creating tuples
point = (3, 4)
rgb = (255, 128, 0)

print(f"Point: {point}")
print(f"Color: {rgb}")

# Access elements
print(f"x = {point[0]}, y = {point[1]}")

# Tuple unpacking — very Pythonic!
x, y = point
print(f"Unpacked: x={x}, y={y}")

r, g, b = rgb
print(f"R={r}, G={g}, B={b}")

# Single-element tuple needs a comma!
single = (42,)    # This IS a tuple
not_tuple = (42)  # This is just 42!
print(f"\n{single} is {type(single).__name__}")
print(f"{not_tuple} is {type(not_tuple).__name__}")

# Try modifying — uncomment to see error:
# point[0] = 10  # TypeError!
```

## Control Flow


### Conditionals — if / elif / else


```python
# Try changing the score!
score = 78

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score {score} → Grade {grade}")

# == vs is: equality vs identity
# Same value, but different objects
a = [1, 2, 3]
b = [1, 2, 3]
print(f"\na == b? {a == b}")  # True (same content)
print(f"a is b? {a is b}")  # False (different objects!)

# Same object (b references a)
b = a
print(f"\nAfter b = a:")
print(f"a is b? {a is b}")  # True (same object)
```

### For Loops & Iteration


```python
fruits = ["apple", "banana", "cherry"]

# ❌ Beginner pattern: manual index tracking
# Verbose, error-prone, less readable
print("Manual index (avoid this):")
for i in range(len(fruits)):
    print(f"  {i}: {fruits[i]}")

# ✅ Pythonic: enumerate() gives index + value
print("\nWith enumerate (prefer this!):")
for i, fruit in enumerate(fruits):
    print(f"  {i}: {fruit}")

# range() examples
print("\nrange(5):", list(range(5)))
print("range(0,10,2):", list(range(0, 10, 2)))

# List comprehension — very Pythonic!
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
evens = [n for n in numbers if n % 2 == 0]
squares = [n**2 for n in numbers]
print(f"\nEvens: {evens}")
print(f"Squares: {squares}")
```

### While Loops & Loop Control


```python
# Simple countdown
count = 5
while count > 0:
    print(f"T-minus {count}...")
    count -= 1
print("Liftoff! 🚀\n")

# break & continue example
print("Finding first even number > 10:")
n = 1
while True:
    n += 1
    if n % 2 != 0:
        continue   # skip odd numbers
    if n > 10:
        print(f"Found it: {n}")
        break       # exit the loop

# Practical: simple convergence check
print("\nSimulated training:")
loss = 1.0
epoch = 0
while loss > 0.01:
    loss *= 0.7  # decay
    epoch += 1
    print(f"  Epoch {epoch}: loss = {loss:.4f}")
print(f"Converged at epoch {epoch}!")
```

## Functions & Sequence Operations


### Defining & Calling Functions


```python
def greet(name, greeting="Hello"):
    """Return a personalized greeting."""
    return f"{greeting}, {name}!"

# Different ways to call
print(greet("Alice"))
print(greet("Bob", "Hi"))
print(greet(greeting="Hey", name="Charlie"))

# Multiple return values
def min_max(numbers):
    """Return both min and max."""
    return min(numbers), max(numbers)

data = [4, 7, 1, 9, 3]
lo, hi = min_max(data)
print(f"\nData: {data}")
print(f"Min: {lo}, Max: {hi}")

# Lambda — anonymous function
square = lambda x: x ** 2
print(f"\nsquare(7) = {square(7)}")

# Lambda with sorted — real-world use
students = [("Alice", 92), ("Bob", 85), ("Charlie", 95)]
ranked = sorted(students, key=lambda s: s[1], reverse=True)
print(f"\nRanking: {ranked}")
```

### Slicing & Negative Indexing


```python
x = [10, 20, 30, 40, 50]

print("Full list:", x)
print("x[1:3]  =", x[1:3])    # [20, 30]
print("x[:3]   =", x[:3])     # [10, 20, 30]
print("x[2:]   =", x[2:])     # [30, 40, 50]
print("x[::2]  =", x[::2])    # [10, 30, 50] every 2nd
print("x[::-1] =", x[::-1])   # reversed!

# Negative indexing
print(f"\nLast item:  x[-1] = {x[-1]}")
print(f"2nd last:   x[-2] = {x[-2]}")
print(f"Last three: x[-3:] = {x[-3:]}")

# String slicing
s = "Python"
print(f"\n'{s}' reversed = '{s[::-1]}'")
print(f"First 3 chars: '{s[:3]}'")

# Practical: extract file extension
# rfind() = find LAST occurrence of substring
# Returns the index, or -1 if not found
filename = "model_weights.h5"
dot_pos = filename.rfind('.')
print(f"\nrfind('.') in '{filename}' → index {dot_pos}")
ext = filename[dot_pos:]
print(f"Extension: {ext}")

# rfind vs find
path = "data/train/images.tar.gz"
print(f"\nfind('.')  → {path.find('.')}")   # first dot
print(f"rfind('.') → {path.rfind('.')}")  # last dot
```

## Classes, Inheritance, and Errors


### Classes & Objects


```python
class Student:
    """Represents a postgraduate student."""

    def __init__(self, name, programme):
        self.name = name
        self.programme = programme
        self.grades = []

    def add_grade(self, score):
        self.grades.append(score)

    def average(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

    def __str__(self):
        return f"{self.name} ({self.programme})"

# Create instances
alice = Student("Alice", "MTech AI")
alice.add_grade(88)
alice.add_grade(92)
alice.add_grade(95)

bob = Student("Bob", "MTech SE")
bob.add_grade(78)
bob.add_grade(85)

print(alice)
print(f"  Average: {alice.average():.1f}")
print(bob)
print(f"  Average: {bob.average():.1f}")
```

### Class Inheritance


```python
class Person:
    def __init__(self, name, year_of_birth):
        self.name = name
        self.year_of_birth = year_of_birth

    def introduce(self):
        return f"I'm {self.name}, born in {self.year_of_birth}"

class Employee(Person):
    def __init__(self, name, year_of_birth, company):
        super().__init__(name, year_of_birth)
        self.company = company

    def introduce(self):  # Override parent method
        base = super().introduce()
        return f"{base}, working at {self.company}"

class Researcher(Employee):
    def __init__(self, name, yob, company, field):
        super().__init__(name, yob, company)
        self.field = field

    def introduce(self):
        base = super().introduce()
        return f"{base}, researching {self.field}"

# Create objects
p = Person("Alice", 1990)
e = Employee("Bob", 1985, "NUS")
r = Researcher("Carol", 1992, "NUS ISS", "NLP")

for obj in [p, e, r]:
    print(obj.introduce())
    print(f"  Is Person? {isinstance(obj, Person)}")
    print()
```

### Error Handling — try / except


```python
# Basic try/except
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("⚠ Cannot divide by zero!")
        return None

print(safe_divide(10, 3))
print(safe_divide(10, 0))

# Handling multiple exception types
def parse_data(raw):
    try:
        value = int(raw)
        return value * 2
    except ValueError:
        print(f"⚠ '{raw}' is not a valid integer")
        return None
    except TypeError:
        print(f"⚠ Expected string, got {type(raw).__name__}")
        return None

print(f"\nparse_data('42') = {parse_data('42')}")
print(f"parse_data('abc') = {parse_data('abc')}")
print(f"parse_data(None) = {parse_data(None)}")

# finally — always executes
print("\nWith finally:")
try:
    data = [1, 2, 3]
    print(data[10])
except IndexError as e:
    print(f"  Caught: {e}")
finally:
    print("  Cleanup done!")
```
