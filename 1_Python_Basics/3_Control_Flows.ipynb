## Control Flow

### Conditionals — if / elif / else

```python
# CONDITIONALS: IF / ELIF / ELSE
score = 78

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score {score} → Grade {grade}")

# == vs is: equality vs identity
# Same value, but different objects
a = [1, 2, 3]
b = [1, 2, 3]
print(f"\na == b? {a == b}")  # True (same content)
print(f"a is b? {a is b}")  # False (different objects!)

# Same object (b references a)
b = a
print(f"\nAfter b = a:")
print(f"a is b? {a is b}")  # True (same object)
```

### For Loops & Iteration

```python
# FOR LOOPS & ITERATION

fruits = ["apple", "banana", "cherry"]

# ❌ Beginner pattern: manual index tracking
# Verbose, error-prone, less readable
print("Manual index (avoid this):")
for i in range(len(fruits)):
    print(f"  {i}: {fruits[i]}")

# ✅ Pythonic: enumerate() gives index + value
print("\nWith enumerate (prefer this!):")
for i, fruit in enumerate(fruits):
    print(f"  {i}: {fruit}")

# range() examples
print("\nrange(5):", list(range(5)))
print("range(0,10,2):", list(range(0, 10, 2)))

# List comprehension — very Pythonic!
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
evens = [n for n in numbers if n % 2 == 0]
squares = [n**2 for n in numbers]
print(f"\nEvens: {evens}")
print(f"Squares: {squares}")
```

### While Loops & Loop Control

```python
# WHILE LOOPS & LOOP CONTROL

# Simple countdown
count = 5
while count > 0:
    print(f"T-minus {count}...")
    count -= 1
print("Liftoff! 🚀\n")

# break & continue example
print("Finding first even number > 10:")
n = 1
while True:
    n += 1
    if n % 2 != 0:
        continue   # skip odd numbers
    if n > 10:
        print(f"Found it: {n}")
        break       # exit the loop

# Practical: simple convergence check
print("\nSimulated training:")
loss = 1.0
epoch = 0
while loss > 0.01:
    loss *= 0.7  # decay
    epoch += 1
    print(f"  Epoch {epoch}: loss = {loss:.4f}")
print(f"Converged at epoch {epoch}!")
```

#### Exercise 1: Conditionals
Given a variable `score`, write an `if / elif / else` block that prints:
- `"Excellent"` if the score is 90 or above
- `"Good"` if the score is 75 or above
- `"Needs Improvement"` otherwise

```python
# EXERCISE 1: CONDITIONALS

score = 82

# Write if / elif / else block that prints:
# "Excellent" - score > 90
# "Good" - score > 75
# "Needs Improvement" - other scores

if score > 90:
    print("Excellent")
elif score > 75:
    print("Good")
else:
    print("Needs Improvement")
```

#### Exercise 3: For Loop and Iteration

#### Exercise 4: Accumulation with a For Loop
Use a `for` loop to calculate the total sum of all numbers in the list.

```python
# EXERCISE 4: ACCUMULATION WITH FOR LOOP

numbers = [3, 7, 11, 2, 9]
total = 0

# Use for loop to calculate total sum of all nums in list
for n in numbers:
    total += n

print(total)
```

#### Exercise 5: While Loop Countdown
Use a `while` loop to print the numbers from 5 down to 1.

```python
# EXERCISE 5: WHILE LOOP COUNT DOWN

count = 5

# Use while loop to print numbers from 5 down to 1
print("Count down:")
while count > 0:
    print(count)
    count -= 1
```

#### Exercise 6: While Loop with Loop Control
Use a `while` loop to print numbers from 1 to 10, but:
- skip printing 4
- stop the loop completely when the number reaches 8

```python
# EXERCISE 6: WHILE LOOP WITH LOOP CONTROL

num = 0

print("Printing numbers:")
while num <= 10:
    if num == 4:
        num += 1
        continue
    elif num == 8:
        break
    print(num)
    num += 1
```
