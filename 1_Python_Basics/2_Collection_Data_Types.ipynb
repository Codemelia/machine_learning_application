```python
# Creating a list
students = [ "Alice", "Bob", "Charlie" ]
print("Original:", students)

# Append and insert
students.append("Diana")
students.insert(2, "Eve")
print("After add:", students)

# Access by index (0-based)
print(f"First: {students[0]}")
print(f"Last: {students[-1]}")

# Modify in place (mutable)
students[0] = "Alicia"
print("Modified:",students)

# List of mixed types
record = [ "John", 36, 1.75, True ]
print(f"\nMixed list: {record}")
print(f"Length: {len(record)}")
```

```python
# TUPLES - IMMUTABLE ORDERED COLLECTIONS

# Create tuples
point = (3, 4)
rgb = (255, 128, 0)
print(f"Point: {point}")
print(f"Color: {rgb}")

# Access elements
print(f"x = {point[0]}, y = {point[1]}")
```
