```python
# Variables and Primitive Data Types

x = 42
print(f"x = {x}, type: {type(x).__name__}")

x = 3.14
print(f"x = {x}, type: {type(x).__name__}")

x = "Hello, World!"
print(f"x = {x}, type: {type(x).__name__}")

x = True
print(f"x = {x}, type: {type(x).__name__}")

x = None
print(f"x = {x}, type: {type(x).__name__}")
```

```python
# Primitive Data Types

a, b = 10, 3

print("a =", a, ", b =", b)
print("Integer division:", a // b)
print("Float division:", a / b)
print("Modulus:", a % b)

# Type Conversion
c = str(a)
print(f"\nstr({a}) = '{c}' (type: {type(c).__name__})")

# Boolean Context
print(f"bool(0) = {bool(0)}")
print(f"bool(1) = {bool(1)}")
print(f"bool('') = {bool('')}")
print(f"bool('x') = {bool('x')}")
print(f"bool(None) = {bool(None)}")

# Arbitrary precision
print(f"\n2^100 = {2**100}")
```

```python
# String Operations

name = "Machine Learning"

#f-string formatting
print(f"Course: {name}")
print(f"Upper: {name.upper()}")
print(f"Length: {len(name)}")

# Splitting words
words = name.split(" ")
print(f"Words: {words}")

# Display character
print(f"First char: '{name[0]}'")
print(f"Last char: '{name[-1]}'")

# Check substring
print(f"'Learn in name? {'Learn' in name}")

# Multi-line strings
bio = """This is
a multi-line
string"""
print(f"\nMulti-line:\n{bio}")
```

```python
# EXERCISE 1: VARIABLES & PRIMITIVE TYPES

age = 21
height = 163.3
student_name = 'Amelia'
is_enrolled = True
middle_name = None

print(age, type(age).__name__)
print(height, type(height).__name__)
print(student_name, type(student_name).__name__)
print(is_enrolled, type(is_enrolled).__name__)
print(middle_name, type(middle_name).__name__)
```
