## Classes, Inheritance, and Errors

### Classes & Objects

```python
# CLASSES & OBJECTS

class Student:
    """Represents a postgraduate student."""

    def __init__(self, name, programme):
        self.name = name
        self.programme = programme
        self.grades = []

    def add_grade(self, score):
        self.grades.append(score)

    def average(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

    def __str__(self):
        return f"{self.name} ({self.programme})"

# Create instances
alice = Student("Alice", "MTech AI")
alice.add_grade(88)
alice.add_grade(92)
alice.add_grade(95)

bob = Student("Bob", "MTech SE")
bob.add_grade(78)
bob.add_grade(85)

print(alice)
print(f"  Average: {alice.average():.1f}")
print(bob)
print(f"  Average: {bob.average():.1f}")
```

### Class Inheritance

```python
# CLASS INHERITANCE

class Person:
    def __init__(self, name, year_of_birth):
        self.name = name
        self.year_of_birth = year_of_birth

    def introduce(self):
        return f"I'm {self.name}, born in {self.year_of_birth}"

class Employee(Person):
    def __init__(self, name, year_of_birth, company):
        super().__init__(name, year_of_birth)
        self.company = company

    def introduce(self):  # Override parent method
        base = super().introduce()
        return f"{base}, working at {self.company}"

class Researcher(Employee):
    def __init__(self, name, yob, company, field):
        super().__init__(name, yob, company)
        self.field = field

    def introduce(self):
        base = super().introduce()
        return f"{base}, researching {self.field}"

# Create objects
p = Person("Alice", 1990)
e = Employee("Bob", 1985, "NUS")
r = Researcher("Carol", 1992, "NUS ISS", "NLP")

for obj in [p, e, r]:
    print(obj.introduce())
    print(f"  Is Person? {isinstance(obj, Person)}")
    print()
```

### Error Handling — try / except

```python
# ERROR HANDLING - TRY / EXCEPT
# Basic try/except
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("⚠ Cannot divide by zero!")
        return None

print(safe_divide(10, 3))
print(safe_divide(10, 0))

# Handling multiple exception types
def parse_data(raw):
    try:
        value = int(raw)
        return value * 2
    except ValueError:
        print(f"⚠ '{raw}' is not a valid integer")
        return None
    except TypeError:
        print(f"⚠ Expected string, got {type(raw).__name__}")
        return None

print(f"\nparse_data('42') = {parse_data('42')}")
print(f"parse_data('abc') = {parse_data('abc')}")
print(f"parse_data(None) = {parse_data(None)}")

# finally — always executes
print("\nWith finally:")
try:
    data = [1, 2, 3]
    print(data[10])
except IndexError as e:
    print(f"  Caught: {e}")
finally:
    print("  Cleanup done!")
```

#### Exercise 1: Defining a Class
Create a class `Car` with:
- attributes `brand` and `year`
- a method `describe()` that returns a string like:

`"Toyota (2024)"`

```python
# EXERCISE 1: DEFINE CLASS

# Create class Car with attributes brand, year & method describe()
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    def describe(self):
        return f"{self.brand} ({self.year})"

my_car = Car("Toyota", 2024)
print(my_car.describe())
```

#### Exercise 2: Class with a Calculation Method
Create a class `Circle` with:
- an attribute `radius`
- a method `area()` that returns `3.14 * radius * radius`

```python
# EXERCISE 2: CLASS WITH CALCULATION METHOD

# Create class Circle with attribute radius & method area()
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius**2)

shape = Circle(5)
print(shape.area())
```

#### Exercise 3: Inheritance
Create a subclass `Manager` that inherits from `Person`.

Add:
- a new attribute `department`
- a method `introduce()` that returns a sentence including the name and department

```python
# EXERCISE 3: INHERITANCE

# Create subclass Manager that inherits from Person
# New attribute department and method introduce()
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"I'm {self.name}, I'm {self.age} years old"

class Manager(Person):
    def __init__(self, name, age, department):
        super().__init__(name, age)
        self.department = department

    def introduce(self):
        base = super().introduce()
        return f"{base}, and I'm from the {self.department} deparatment"

mgr = Manager("Alice", 35, "Analytics")
print(mgr.introduce())
```
