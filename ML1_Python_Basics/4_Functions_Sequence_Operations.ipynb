## Functions & Sequence Operations

### Defining & Calling Functions

```python
# DEFINING & CALLING FUNCTIONS

def greet(name, greeting="Hello"):
    """Return a personalized greeting."""
    return f"{greeting}, {name}!"

# Different ways to call
print(greet("Alice"))
print(greet("Bob", "Hi"))
print(greet(greeting="Hey", name="Charlie"))

# Multiple return values
def min_max(numbers):
    """Return both min and max."""
    return min(numbers), max(numbers)

data = [4, 7, 1, 9, 3]
lo, hi = min_max(data)
print(f"\nData: {data}")
print(f"Min: {lo}, Max: {hi}")

# Lambda — anonymous function
square = lambda x: x ** 2
print(f"\nsquare(7) = {square(7)}")

# Lambda with sorted — real-world use
students = [("Alice", 92), ("Bob", 85), ("Charlie", 95)]
ranked = sorted(students, key=lambda s: s[1], reverse=True)
print(f"\nRanking: {ranked}")
```

### Slicing & Negative Indexing

```python
# SLICING & NEGATIVE INDEXING

x = [10, 20, 30, 40, 50]

print("Full list:", x)
print("x[1:3]  =", x[1:3])    # [20, 30]
print("x[:3]   =", x[:3])     # [10, 20, 30]
print("x[2:]   =", x[2:])     # [30, 40, 50]
print("x[::2]  =", x[::2])    # [10, 30, 50] every 2nd
print("x[::-1] =", x[::-1])   # reversed!

# Negative indexing
print(f"\nLast item:  x[-1] = {x[-1]}")
print(f"2nd last:   x[-2] = {x[-2]}")
print(f"Last three: x[-3:] = {x[-3:]}")

# String slicing
s = "Python"
print(f"\n'{s}' reversed = '{s[::-1]}'")
print(f"First 3 chars: '{s[:3]}'")

# Practical: extract file extension
# rfind() = find LAST occurrence of substring
# Returns the index, or -1 if not found
filename = "model_weights.h5"
dot_pos = filename.rfind('.')
print(f"\nrfind('.') in '{filename}' → index {dot_pos}")
ext = filename[dot_pos:]
print(f"Extension: {ext}")

# rfind vs find
path = "data/train/images.tar.gz"
print(f"\nfind('.')  → {path.find('.')}")   # first dot
print(f"rfind('.') → {path.rfind('.')}")  # last dot
```

#### Exercise 1: Basic Function
Write a function `square_number(x)` that returns the square of `x`.

```python
# EXERCISE 1: BASIC FUNCTION

# Return square of x
def square_number(x):
    return x**2

print(square_number(4))
print(square_number(9))
```

#### Exercise 2: Function with Multiple Arguments
Write a function `calculate_total(price, quantity)` that returns the total cost.

```python
# EXERCISE 2: FUNCTION WITH MULTIPLE ARGUMENTS

# Return total calculated price
def calculate_total(price, quantity):
    return price * quantity

print(calculate_total(12.5, 4))
print(calculate_total(3, 7))
```

#### Exercise 3: Function with Default Argument
Write a function `introduce(name, role="student")` that returns a string in this format:

`"Alice is a student."`

```python
# EXERCISE 3: FUNCTION WITH DEFAULT ARGUMENT

# Introduce person using format
def introduce(name, role="student"):
    return f"{name} is a {role}"

print(introduce("Alice"))
print(introduce("Bob", "instructor"))
```

#### Exercise 6: Combine Functions and Slices
Write a function `first_and_last(values)` that returns a tuple containing:
- the first element of the sequence
- the last element of the sequence

```python
# EXERCISE 6: COMBINE FUNCTIONS & SLICES

# return tuple with fist & last element
def first_and_last(values):
    return values[0], values[-1]

print(first_and_last([10, 20, 30, 40]))
print(first_and_last("Python"))
```
