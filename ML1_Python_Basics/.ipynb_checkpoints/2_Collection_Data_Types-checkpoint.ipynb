## Collection Data Types

### Lists — Mutable Ordered Collections

```python
# Creating a list
students = [ "Alice", "Bob", "Charlie" ]
print("Original:", students)

# Append and insert
students.append("Diana")
students.insert(2, "Eve")
print("After add:", students)

# Access by index (0-based)
print(f"First: {students[0]}")
print(f"Last: {students[-1]}")

# Modify in place (mutable)
students[0] = "Alicia"
print("Modified:",students)

# List of mixed types
record = [ "John", 36, 1.75, True ]
print(f"\nMixed list: {record}")
print(f"Length: {len(record)}")
```

### Tuples — Immutable Ordered Collections

```python
# Create tuples
point = (3, 4)
rgb = (255, 128, 0)
print(f"Point: {point}")
print(f"Color: {rgb}")

# Access elements
print(f"x = {point[0]}, y = {point[1]}")

# Tuple unpacking
x, y = point
r, g, b = rgb
print(f"Unpacked: x = {x}, y = {y}")
print(f"Unpacked: r = {r}, g = {g}, b = {b}")

# Single-element tuple needs coma
single = (42,) # This IS a tuple
not_tuple = (42) # This is an int type
print(f"\n{single} is {type(single).__name__}")
print(f"{not_tuple} is {type(not_tuple).__name__}")

# Try modifying
# point[0] = 10 #TypeError
```

### Dictionaries — Key-Value Stores

```python
# ML hyperparameters as a dict
config = {
    "learning_rate": 0.001,
    "epochs": 50,
    "batch_size": 32,
    "optimizer": "adam"
}

print("Config:", config)
print(f"LR: {config['learning_rate']}")

# Safe access with get()
print(f"Dropout: {config.get('dropout', 'not set')}")

# Add / update
config["dropout"] = 0.5
config["epochs"] = 100
print(f"\nUpdated: {config}")

# Iterate over key-value pairs
print("\nAll settings:")
for key, value in config.items():
    print(f"{key}: {value}")
```

```python
# EXERCISE 1: LISTS

students = [ "Alice", "Bob", "Charlie" ]

# Change "Bob" to "Bobby" and add "Diana" to end of list
students[1] = "Bobby"
students.append("Diana")

# Print first & last element and lenth of list
print(f"First: {students[0]}")
print(f"Last: {students[-1]}")
print(f"Length: {len(students)}")
```

#### Exercise 4: Dictionaries

#### Exercise 5: Dictionary Iteration
Use a `for` loop with `.items()` to print each key-value pair in the dictionary below in the format:

`key => value`

```python
# EXERCISE 5: DICTIONARY ITERATION

inventory = {
    "pen": 12,
    "notebook": 5,
    "eraser": 9
}

for key, value in inventory.items():
    print(f"Key: {key}, Value: {value}")
```
