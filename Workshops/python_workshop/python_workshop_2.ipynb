Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]

Your solution need not be the most efficient one. Of course you are encouraged to come out with the most efficient solution.

## Complete the implementation

```python
# TODO: return a pair of indices whose values add up to target
# TODO: if nums[i] + nums[j] equals target, return (i, j)

# METHOD 1: DICTIONARY
def find1(nums, target):
    
    # dictionary to store difference to target (key) to current index (value)
    seen = {}
    
    for i, n in enumerate(nums):
        # get difference to target
        diff = target - n

        # if curr num is already in dictionary, that means target is hit
        # return previously stored index and current index
        if n in seen:
            return seen[n], i

        # if difference not in dictionary, add to dictionary
        seen[diff] = i

    return None

# METHOD 2: NESTED FOR LOOP
def find2(nums, target):

    # outer loop
    for i in range(len(nums)):

        # inner loop starting from next index
        for j in range(i + 1, len(nums)):

            # num on outer loop + num on inner loop hits target
            # return outer loop index and inner loop index
            if nums[i] + nums[j] == target:
                return i, j

    return None
```
