## Exercise 1

Write a program to compute gross pay from hours worked and hourly rate.
Take into account that the factory gives the employee 1.5 times the hourly rate for hours worked above 40 hours.

Example:
- Hours = 45
- Rate = 10
- Pay = 475.0


````python
# Starter code
hours = 45
rate = 10

# TODO: compute regular pay and overtime pay
def compute_pay(hours, rate):
    if hours < 0:
        pass
    elif hours > 40:
        return (40 * rate) + ((hours - 40) * rate * 1.5)
    else:
        return hours * rate

pay = compute_pay(hours, rate)
print("Pay:", pay)
````

## Exercise 3

Write a shutting down program:

First, def a function, shut_down, that takes one argument s. 

Then, if the shut_down function receives an s equal to "yes", it should return "Shutting down" 

Alternatively, elif s is equal to "no", then the function should return "Shutdown aborted". 

Finally, if shut_down gets anything other than those inputs, the function should return "Sorry".

````python
def shut_down(s):
    if s == 'yes':
        return "Shutting down"
    elif s == 'no':
        return "Shutdown aborted"
    else:
        return "Sorry"

print(shut_down('yes'))
print(shut_down('no'))
print(shut_down(0))
````

## Exercise 4

Write a function that accepts a sentence and capitalizes the first letter of each word in the string.
You may use the `split()` method to split the sentence into words.


````python
# Starter code
def capitalize_words(sentence):
    # TODO: return the sentence with each word capitalized
    words = [ w.title() for w in sentence.split() ]
    return " ".join(words)

print(capitalize_words("i have a dream."))
````

## Exercise 6

Follow the steps:

1. Create a class `Triangle`. Its `__init__()` method should take `self`, `angle1`, `angle2`, and `angle3` as arguments.
2. Create a variable named `number_of_sides` and set it equal to 3.
3. Create a method named `check_angles`. The sum of a triangle's three angles is 180. It should return `True` if the sum of `self.angle1`, `self.angle2`, and `self.angle3` is 180, and `False` otherwise.
4. Create a variable named `my_triangle` and set it equal to a new instance of your `Triangle` class. Pass it three angles that sum to 180 (e.g. 90, 30, 60).
5. Print out `my_triangle.number_of_sides` and `my_triangle.check_angles()`.


````python
class Triangle:
    number_of_sides = 3

    def __init__(self, angle1, angle2, angle3):
        # TODO: store the three angles
        self.angle1, self.angle2, self.angle3 = angle1, angle2, angle3

    def check_angles(self):
        # TODO: return True if the three angles sum to 180
        return True if self.angle1 + self.angle2 + self.angle3 == 180 else False

my_triangle = Triangle(90, 30, 60)
print(my_triangle.number_of_sides)
print(my_triangle.check_angles())
````

## Exercise 7

Given a list of integers, write a loop that counts how many times each value appears and stores the result in a dictionary.

Use this example list:
`[1, 3, 3, 1, 2]`

Expected result:
`{1: 2, 2: 1, 3: 2}`


````python
# TODO: count each value and store the result in counts
def count_dict(counts, numbers):
    for n in numbers:
        counts[n] = counts.get(n, 0) + 1
    return counts.items()

counts = dict(sorted(count_dict({}, [1, 3, 3, 1, 2])))
print(counts)
````

## Exercise 9

Create a list with the following values: 5, 6, 4, 7, 10.

Answer the following questions:
- a) Sort the array. (Use the `sort` method)
- b) Access the 4th element of the sorted array.
- c) Reverse the order of the array. (Use the `reverse` method)
- d) Remove the value 6 from the array.
- e) Delete the third element.
- f) Delete the first two elements


````python
values = [5, 6, 4, 7, 10]
print("Original:", values)

# TODO: perform parts (a) to (f)
# sort array
values = sorted(values)
print("Sorted:", values)

# access fourth element of array
print("Element 4:", values[3])

# reverse order of array
values = list(reversed(values))
print("Reversed:", values)

# remove value 6 from array
values.remove(6)
print("Removed value 6:", values)

# delete third element
values.pop(2)
print("Removed element 3:", values)

# delete first two elements
values = values[-1::]
print("Removed first 2 elements:", values)
````
