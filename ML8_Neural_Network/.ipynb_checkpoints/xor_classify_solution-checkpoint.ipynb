```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# ---------------------------------------------------
# Set random seeds
# ---------------------------------------------------
# This makes the training result reproducible.
# Every time you run the code, the initialized weights
# and training behavior will be similar.
# ---------------------------------------------------
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# ---------------------------------------------------
# XOR Dataset
# ---------------------------------------------------
# Input:
#   2 binary features
#
# Output:
#   XOR truth table result
#
# XOR logic:
#   same values   -> 0
#   different     -> 1

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=np.float32)

# Labels / target values
# This is a binary classification problem.
# The network predicts either:
#   class 0
#   class 1

y = np.array([
    [0],
    [1],
    [1],
    [0]
], dtype=np.float32)
```

```python
# ---------------------------------------------------
# Build Neural Network
# ---------------------------------------------------
# Sequential model:
#   layers are stacked one after another
#
# Architecture:
#
# Input layer:
#   2 input values
#
# Hidden layer:
#   4 neurons
#   activation = tanh
#
# Output layer:
#   1 neuron
#   activation = sigmoid
#
# Why hidden layer?
# ---------------------------------------------------
# XOR is NOT linearly separable.
#
# A single linear layer cannot solve XOR.
#
# Hidden layers + non-linear activation functions
# allow the network to learn non-linear boundaries.
# ---------------------------------------------------

model = keras.Sequential([

    # Input layer
    # shape=(2,)
    # means:
    # each training sample contains 2 features
    layers.Input(shape=(2,)),

    # Hidden layer
    #
    # Dense:
    # fully connected layer: 4 neurons with tanh activation
     # Non-linearity is important for XOR.
    layers.Dense(8, activation="tanh"),

    # Output layer
    # 1 neuron because binary classification
    # sigmoid activation:
    # converts output into probability
    layers.Dense(1, activation="sigmoid")
])
```

```python
# ---------------------------------------------------
# Compile Model
# ---------------------------------------------------
# Configure training settings.
#
# optimizer:
#   Adam optimizer
#   handles weight updates during gradient descent
#
# learning_rate:
#   controls step size during optimization
#
# loss:
#   binary_crossentropy
#
#   Standard loss function for binary classification.
#
# metrics:
#   accuracy
#
#   Percentage of correct predictions.
# ---------------------------------------------------

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.05),

    # Binary classification loss
    loss="binary_crossentropy",

    # Track accuracy during training
    metrics=["accuracy"]
)
```

```python
# ---------------------------------------------------
# Train Model
# ---------------------------------------------------
# model.fit():
#   trains the neural network
# epochs:
#   number of full training passes
# Since XOR dataset is tiny,
# we train many epochs.
#
# verbose=0:
#   suppress training logs
# ---------------------------------------------------

model.fit(X, y, epochs=1000, verbose=0)

# ---------------------------------------------------
# Evaluate Model
# ---------------------------------------------------
# Check final training performance.
#
# loss:
#   prediction error
#
# accuracy:
#   percentage of correct predictions
# ---------------------------------------------------

loss, accuracy = model.evaluate(X, y, verbose=0)

print(f"Training loss: {loss:.4f}")
print(f"Training accuracy: {accuracy:.4f}")
```

```python
# ---------------------------------------------------
# Make Predictions
# ---------------------------------------------------
# model.predict():
#   returns probabilities
#
# Example:
#   0.98
#   0.03
#
# Convert probabilities into classes:
#
# >= 0.5 -> class 1
# <  0.5 -> class 0
# ---------------------------------------------------

probabilities = model.predict(X)

predicted_classes = (probabilities >= 0.5).astype(int)

# ---------------------------------------------------
# Display Results
# ---------------------------------------------------
# Show:
#   input
#   true label
#   predicted probability
#   predicted class
# ---------------------------------------------------

for input_value, true_label, prob, pred in zip(
    X,
    y,
    probabilities,
    predicted_classes
):
    print(
        f"Input: {input_value.astype(int)} | "
        f"True: {int(true_label[0])} | "
        f"Probability: {prob[0]:.4f} | "
        f"Predicted class: {pred[0]}"
    )
```
